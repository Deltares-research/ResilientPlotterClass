## Creating the ResilientPlotterClass logo

### Packages

```python
import matplotlib.font_manager as fm
from matplotlib import patheffects

from resilientplotterclass import rpc
```

### Plot logo

```python
# Plot logo with text
font_path = "DMSans-ExtraBold.ttf"
font = fm.FontProperties(fname=font_path)

fig, ax = rpc.subplots(figsize=(13.5, 2.5))

path_effects = [patheffects.withStroke(linewidth=5, foreground="#FDC372")]
ax.text(0.05, 0.3, "RPCLASS", fontsize=150, fontproperties=font, va="center", color="#FDC372", path_effects=path_effects, zorder=0.5)

path_effects = [patheffects.withStroke(linewidth=5.5, foreground="#3C94B9")]
ax.text(0.05, 0.3, "RPC", fontsize=150, fontproperties=font, va="center", color="#3C94B9", path_effects=path_effects, zorder=0.5)

ax.set_axis_off()
rpc.save(fig=fig, file_path="rpc_logo.png", transparent=True)
```
